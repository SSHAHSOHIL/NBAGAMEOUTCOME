```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
```

```python
test = games[games["Date"] > '2024-02-01']
```

```python
predictors = ["Home/Away_code", "Opp_Code", "Day_Code", "ELO"]
```

```python
rf.fit(train[predictors], train["Target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
grouped_games = games.groupby("Team")
```

```python
group = grouped_games.get_group("DAL")
```

```python
group
```

```python
def rolling_averages(team, cols, col_new):
    team = team.sort_values("Date")
    rolling_stats = team[cols].rolling(3, closed = "left").mean()
    team[col_new] = rolling_stats
    team = team.dropna(subset=col_new)
    return team
```

```python
cols = ["PTS", "3PM", "3P%", "REB", "AST%", "OREB%", "DREB%", "PACE", "ELO"]
new_cols = [f"{c}_rolling" for c in cols]
```

```python
new_cols
```

```python
rolling_averages(group, cols, new_cols)
```

```python
group.dtypes
```

```python
games_rolling = games.groupby("Team").apply(lambda x: rolling_averages(x, cols, new_cols))
```

```python
games_rolling
```

```python
games_rolling = games_rolling.droplevel("Team")
```

```python
games_rolling
```

```python
games_rolling.index = range(games_rolling.shape[0])
```

```python
games_rolling
```

```python
def predictions(values, predictors):
    train = values[values["Date"] < '2024-02-01']
    test = values[values["Date"] > '2024-02-01']
    rf.fit(train[predictors], train["Target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["Target"], predicted=preds), index=test.index)
    precision = precision_score(test["Target"],preds)
    return combined, precision
```

```python
combined, precision = predictions(games_rolling, predictors + new_cols)
```
